##### (i) Age vs Spending

```python
# Create crosstab to determine correlation between the age and spending_flag

crosstab_age_spending = pd.crosstab(df_final['age'], df_final['spending_flag'], dropna = False)
```

```python
crosstab_age_spending
```

```python
# Create age group grouping using if-statement with For-loops

age_group = []

for value in df_final['age']:
    if value < 20:
        age_group.append('18-19')
    elif (value >= 20) and (value <= 29):
        age_group.append('20-29')
    elif (value >= 30) and (value <= 39):
        age_group.append('30-39')
    elif (value >= 40) and (value <= 49):
        age_group.append('40-49')
    elif (value >= 50) and (value <= 59):
        age_group.append('50-59')
    elif (value >= 60) and (value <= 69):
        age_group.append('60-69')
    elif (value >= 70) and (value <= 79):
        age_group.append('70-79')
    else:
        age_group.append('80+')
```

```python
# Create new age_group column

df_final['age_group'] = age_group
```

#### (v) “Single adult” or “Young parent”

* Characteristic of "Single adult" is n_dependants = 0
* Characteristics of "Young parent" are age <= 29 and n_dependant = 1,2,3

```python
# Create new parental_status column based on above criteria

df_final.loc[df_final['n_dependants'] == 0, 'parental_status'] = 'Single adult'
df_final.loc[(df_final['n_dependants'] == 1) & (df_final['age'] <=29), 'parental_status' ] = 'Young parent'
df_final.loc[(df_final['n_dependants'] == 2) & (df_final['age'] <=29), 'parental_status' ] = 'Young parent'
df_final.loc[(df_final['n_dependants'] == 3) & (df_final['age'] <=29), 'parental_status' ] = 'Young parent'
```

#### (vii) Frequency and classification of order_hour_of_day

```python
df_final['order_hour_of_day'].value_counts()
```

##### order_hour_of_day classification:
* Morning = 6-11
* Afternoon = 12-17
* Evening = 18-23
* After midnight = 24-5

```python
# Classification of the order_hour_of_day

hour_classification=[]

for value in df_final['order_hour_of_day']:
    if (value >= 6) and (value <= 11):
        hour_classification.append('Morning')
    elif (value >= 12) and (value <= 17):
        hour_classification.append('Afternoon')
    elif (value >= 18) and (value <= 23):
        hour_classification.append('Evening')
    else:
        hour_classification.append('After midnight')
```

```python
# Create new hour_classification column
df_final['hour_classification'] = hour_classification
```

```python
df_final['hour_classification'].value_counts(dropna=False)
```
