##### Project Brief Q1. What are the busiest days of the week and hours of the day? (i.e., the days and times with the most orders) 

'day_of_week' and 'orders_day_of_week' are contains the same data information. The only difference is:
* 'day_of_week' has data in the name of the days (i.e Monday, Tuesday, etc)
* 'orders_day_of_week' has data in numerical form

```python
# Frequency bar chart of total transaction orders during days of the week

bar1_1 = sns.barplot(x=df_final['day_of_week'].value_counts().index, 
                    y=df_final['day_of_week'].value_counts(),
                    color = 'c', 
                    order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).set(
                    title="Total Daily Transactions")

plt.xticks(rotation=45)
plt.xlabel('Days of Week')
plt.ylabel('Number of Transactions (million)')

plt.savefig('Frequency_Orders_Day_of_Week.png')
```

```python
# Frequency of 'day_of_week' column in descending order

df_final['day_of_week'].value_counts(dropna = False)
```

```python
# Frequency bar chart of total transaction orders during hours of the day

bar1_2 = sns.barplot(x=df_final['order_hour_of_day'].value_counts().index, 
                    y=df_final['order_hour_of_day'].value_counts(),
                    color = 'c').set(
                    title="Total Hourly Transactions")

plt.xlabel('Hour of Day')
plt.ylabel('Number of Transactions (million)')

plt.savefig('Frequency_Orders_Hour_of_Day.png')
```

```python
# Frequency of 'order_hour_of_day' column in descending order

df_final['order_hour_of_day'].value_counts()
```

##### Project Brief Q2. Are there particular times of the day when people spend the most money? (as this might inform the type of products they advertise at these times.)


```python
# Grouping sum of 'total_spending' column by 'order_hour_of_day' column.

sum_tot_spending = df_final.groupby('order_hour_of_day').agg({'total_spending':['sum']})
```

```python
sum_tot_spending
```

```python
# Sorting by descending order

sort_sum_tot_spending = sum_tot_spending.apply(lambda x: x.sort_values(ascending=False))
```

```python
sort_sum_tot_spending
```

```python
# Create line chart for total spending vs hour of day

sum_tot_spending.plot(kind = 'line', color = 'c')
plt.title('Total Hourly Spending')
plt.xlabel('Hour of Day')
plt.ylabel('Spending ($)')
plt.xticks(rotation=0)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.savefig('Total Hourly Spending.png')
```

```python
# Checking average spending vs hour of day

avg_spending2 = df_final.groupby('order_hour_of_day')['total_spending'].mean()
```

```python
avg_spending2
```

```python
# Create line chart for Average Spending vs Hour of Day
avg_spending2.plot(kind = 'line', color = 'c')
plt.title('Average Spending vs Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Spending ($)')
plt.xticks(rotation=0)
plt.legend('')
```
