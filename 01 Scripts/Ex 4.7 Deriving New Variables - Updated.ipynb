#### (i) 'price_label' column - Using User-Defined Function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return 'Not enough data'
```

```python
# Use the dataframe to specify the appropriate string label to return for each product.

df['price_range'] = df.apply(price_label, axis=1)
```

#### (ii) 'Busiest day' column - Using If-Statements with For Loops

```python
# Frequency of the “orders_day_of_week” column

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Using for-loop to create new labels for busiest day 

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

#### 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Using for-loop to create new labels for busiest days 

result2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result2.append("Busiest days")
  elif value == 1:
    result2.append("Busiest days")
  elif value == 4:
    result2.append("Slowest days")
  elif value == 3:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
# Create new column 'busiest_days' to contain these labels

df_ords_prods_merged['busiest_days'] = result2
```

```python
# Value counts for each label

df_ords_prods_merged['busiest_days'].value_counts(dropna = False)
```

#### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

#### 5. Print the frequency for this new column.

```python
# Identify busiest hour - by determining the frequency of the “orders_day_of_week” column

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

10am is the busiest hour of the day, while 3 am is the least busiest hour of the day.

For the following exercise, I will classify 10-19th hour as the busiest, while 2-5th hour as the least busiest.

```python
# Using for-loop to create new labels for the time period of orders 

result_hour = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value >= 10 & value <= 19:
    result_hour.append("Most orders")
  elif value >= 2 & value <= 5:
    result_hour.append("Fewest orders")
  else:
    result_hour.append("Average orders")
```

```python
# Print result_hour

result_hour
```

```python
# Create a new column containing these labels called “busiest_period_of_day.”

df_ords_prods_merged['busiest_period_of_day'] = result_hour
```
