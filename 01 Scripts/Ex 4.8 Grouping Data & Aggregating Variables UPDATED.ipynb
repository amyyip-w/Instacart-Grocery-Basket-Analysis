#### 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
* If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
* If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merged.groupby('user_id').agg({'prices': ['mean', 'median', 'min', 'max']}).head(5)
```

The prices of products purchased by loyal customers are only slightly lower on average. The minimum and maximum range of product prices purchased are the same across the loyalty categories.

```python
# Create a new 'avg_spending' column for average spending of each user

ords_prods_merged['avg_spending'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

ords_prods_merged.loc[ords_prods_merged['avg_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

ords_prods_merged.loc[ords_prods_merged['avg_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Print the frequency of new “spending_flag” column using the value_counts() function:

ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged[['user_id', 'avg_spending', 'spending_flag']].head()
```

#### 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
* If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
* If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
* If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a new 'median_days_since_prior_order' column

ords_prods_merged['median_days_since_prior_order'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

ords_prods_merged.loc[ords_prods_merged['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

ords_prods_merged.loc[(ords_prods_merged['median_days_since_prior_order'] > 10) & (ords_prods_merged['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

ords_prods_merged.loc[ords_prods_merged['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
