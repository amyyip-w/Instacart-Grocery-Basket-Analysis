# 05. Create a subset using top 5 departments: Produce, Dairy eggs, Snacks, Beverages, Frozen.

```python
# Frequency of 'department' column in descending order

df_final['department'].value_counts()
```

Top 5 departments in terms of number of transaction orders: Produce, Dairy eggs, Snacks, Beverages, Frozen.

```python
# Create a flag for top 5 departments using If-Statements with For-Loops. 

result = []

for value in df_final['department']:
    if value == 'Produce':
        result.append('Top')
    elif value == 'Dairy eggs':
        result.append('Top')
    elif value == 'Snacks':
        result.append('Top')
    elif value == 'Beverages':
        result.append('Top')
    elif value == 'Frozen':
        result.append('Top')
    else:
        result.append('NaN')
```

```python
df_final['top_5_departments'] = result
```

# 06. Popular products - Create flag for top 10 selling products

```python
total_products_sold = df_final.groupby('product_name').agg({'order_numbers_users_made': ['sum']})
```

```python
# Sort products by by descending order

sort_total_products_sold = total_products_sold.apply(lambda x: x.sort_values(ascending=False))
```

```python
# Top 10 selling products

sort_total_products_sold.head(10)
```
